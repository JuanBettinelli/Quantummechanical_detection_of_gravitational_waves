```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
N = 4
Nlist =[]
for l in range(0,N):
    Nlist.append(l)

sz = []
sx = []
w = []
x = []
p = []
for i in Nlist: #range(0, (N-1)):
    sz.append(sigmaz()) 
    sx.append(sigmax()) 
    w.append(1.0*2*np.pi) 
    x.append(i*1*10**-9)
    
tlist = np.linspace(0,5,200)

c = 3**8
E0 = 10**-3
W = 0.5*10**5

psi0 = basis(2,0)
e_ops = [sigmax(), sigmay(), sigmaz()]
```

```python
Ht = []
H0 = 0
E = []
NL = [1,2]

for j in Nlist: 
    H0 = H0 -w[j]*sz[j]/N

for k in Nlist:
    if k>0:
        E.append((E0/(N-1))*(np.cos(W*x[k-1]/c-W*tlist)-np.cos(W*x[k]/c-W*tlist)+1*10**-9))
        Ht.append(sx[j-1]*sx[j])

    
#H = [H0,[Ht[0],E[0]],[Ht[1],E[1]]]
    
 
H = [H0]
for k in range(0, N-1):
    H.append([Ht[k], E[k]])


H
```

```python
output = mesolve(H, psi0, tlist, [], e_ops) 
```

```python
plot_expectation_values(output);
```

```python
sxo = output.expect[0]
syo = output.expect[1]
szo = output.expect[2] 

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(tlist, np.real(sxo), 'r')
ax.plot(tlist, np.real(syo), 'b')
ax.plot(tlist, np.real(szo), 'g')
```
