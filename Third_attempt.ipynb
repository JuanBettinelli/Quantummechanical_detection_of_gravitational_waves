```python
import matplotlib.pyplot as plt
```

```python
import numpy as np   #downloding Maths tools
from math import pi, cos #donloding functions
```

```python
from qutip import * #downloding Qunatum tool box
```

```python
sz1 = sigmaz() #declre variables, Pauli Matrx
sz2 = sigmaz()
sx1 = sigmax()
sx2 = sigmaz()
w1 = 0.01*2*pi #angular momentum
w2 = 0.01*2*pi
tlist = np.linspace(0,50,2000)  #describe a time list
f = 10**2 # binary star frequncy
G = 6.67*10**-11 #m/kg^2  #gravitational constant
c = 3*10**8 #m/s  speed of light
i = 0 #orbit inclination
m0 = 2*10*30 #kg   mass of sun
m1 = 2.4*m0 #kg    mass of star1
m2 = 2.4*m0 #kg    mass of star2
r = 3*10**9 #m     distance to system
```

```python
psi0 = basis(2,0)  #initial state
e_ops = [sigmax(), sigmay(), sigmaz()]   #operators to be evolved in master equation solver
```

```python
mu = (m1*m2)/(m1+m2)   #defining amletude of grav. wave, reduced mass of system
M = m1 + m2  #total mas of system
h1=(2*G**(5/3)/c**4)
h2=(1+np.cos(i)**2)
h3=(mu/r)*(np.pi*M*f)**(2/3)
h4=np.cos(2*np.pi*f*tlist)
hplus=h1*h2*h3*h4 #h plus polarisation

h5 = (4*G**(5/3)/c**4)
h6 = np.cos(i)
h7 = (mu/r)*(np.pi*M*f)**(2/3)
h8 = np.sin(2*np.pi*f*tlist)
hx = h5*h6*h7*h8 #h cros polariastion

hsqu = (10**-21*np.sin(f*tlist))**1 #hx**2 + hplus**2  # no units   square of ampletude
```

```python
Et0 = (np.pi*c**3)/(4*G)*f**2*hsqu   #Time dependent value of hamilonion
Et =  4*np.pi*Et0/r**2  #units Wm^-2
```

```python
H0 = -w1*sz1/2 -w2*sz2/2   #time independent part of hamiltonian
H1 =  sx1*sx2    # part of the time dependent part of the hamiltonian

H = [H0,[H1,Et]]   #compleate time dependet hamilatioan using list
```

```python
output = mesolve(H, psi0, tlist, [], [sigmax(), sigmay(), sigmaz()]) #master equation solver
```

```python
#plt.axis([0, 10, -0.5, 1.1])
#axes.plot(tlist, output.expect[2], label=r'$\left<\sigma_z\right>$')
fig, axes = plt.subplots(figsize=(12,6))   # plot of expectation values
axes.plot(tlist, output.expect[1], label=r'$\left<\sigma_y\right>$')
axes.plot(tlist, output.expect[0], label=r'$\left<\sigma_x\right>$')

axes.set_xlabel(r'$t/s$', fontsize=20)
axes.legend(loc=2);
```

```python

FFT0 = np.fft.rfft(output.expect[0])
FFT1 = np.fft.rfft(output.expect[1])
FFT2 = np.fft.rfft(output.expect[2])


fig, axes = plt.subplots(figsize=(12,6))   # plot of expectation values
axes.plot( abs(FFT0), label=r'$\left<\sigma_y\right>$')
axes.plot(abs(FFT1), label=r'$\left<\sigma_x\right>$')
axes.plot(abs(FFT2), label=r'$\left<\sigma_x\right>$')

axes.set_xlabel(r'$f/Hz$', fontsize=20)
axes.legend(loc=2);
```

```python
HI = -w1*sz1/2 -w2*sz2/2 + sx1*sx2  

result = mesolve(HI, psi0, tlist, [], e_ops) 

fig, axes = plt.subplots(figsize=(12,6))
axes.plot(tlist, result.expect[1], label=r'$\left<\sigma_y\right>$')
axes.plot(tlist, result.expect[0], label=r'$\left<\sigma_x\right>$')

axes.set_xlabel(r'$t/s$', fontsize=20)
axes.legend(loc=2);
```
