```python
import numpy as np
from math import pi, cos
```

```python
from qutip import *
```

```python
sz1 = sigmaz()
sz2 = sigmaz()
sx1 = sigmax()
sx2 = sigmaz()
w1 = 1.0*2*pi
w2 = 1.0*2*pi
tlist = np.linspace(0,5,200)
#E0 = 10**-3
f = 10**-3
G = 6.67*10**-11
c = 3*10**8
i = 0
m0 = 2*10*30
m1 = 1.4*m0
m2 = 1.4*m0
r = 3*10**20
```

```python
psi0 = basis(2,0)
e_ops = [sigmax(), sigmay(), sigmaz()]
```

```python
mu = (m1*m2)/(m1+m2)
M = m1 + m2
h1=(2*G**(5/3)/c**4)
h2=(1+np.cos(i)**2)
h3=(mu/r)*(np.pi*M*f)**(2/3)
h4=np.cos(2*np.pi*f*tlist)
h=h1*h2*h3*h4
```

```python
Et0 = (np.pi*c**3)/(4*G)*f**2*h**2
Et = 4*np.pi*r**2*Et0
Et
```

```python
H0 = -w1*sz1/2 -w2*sz2/2
H1 =  sx1*sx2
#E = E0*np.cos(-W*tlist)
H = [H0,[H1,Et]]
H
```

```python
output = mesolve(H, psi0, tlist, [], e_ops) 
```

```python
plot_expectation_values(output);
```

```python
sxo = output.expect[0]
syo = output.expect[1]
szo = output.expect[2] 

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(tlist, sxo, 'r')
ax.plot(tlist, syo, 'b')
#ax.plot(tlist, szo, 'g')
```
