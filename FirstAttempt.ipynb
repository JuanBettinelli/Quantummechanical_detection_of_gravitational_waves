```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from math import pi, cos
```

```python
from qutip import *
```

```python
sz1 = sigmaz()
sz2 = sigmaz()
sx1 = sigmax()
sx2 = sigmaz()
w1 = 1.0*2*pi
w2 = 1.0*2*pi
tlist = np.linspace(0,5,200)
E0 = 10**-3
W = 10**5
```

```python
psi0 = basis(2,0)
e_ops = [sigmax(), sigmay(), sigmaz()]
```

```python
H0 = -w1*sz1/2 -w2*sz2/2
H1 =  sx1*sx2
E = E0*np.cos(-W*tlist)
H = [H0,[H1,E]]
```

```python
output = mesolve(H, psi0, tlist, [], e_ops) 
```

```python
plot_expectation_values(output);
```

```python
sxo = output.expect[0]
syo = output.expect[1]
szo = output.expect[2] 

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(tlist, sxo, 'r')
ax.plot(tlist, syo, 'b')
ax.plot(tlist, szo, 'g')
```
