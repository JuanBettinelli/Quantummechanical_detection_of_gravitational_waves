```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
h = 10**-21
D = 10**5
c = 3*10**8
R = 20
L = 1*10**-3
By = 1
WV = D/c
x = 0
N = 2000
tlist = np.linspace(0, 50000, N)
V0 = 0
SX = 1
SZ = 1
```

```python
B = []
for t in tlist:
    b = 1/2*By*(h*WV*np.sin(WV*(t-x)))/(-1-h*np.cos(WV*(t-x)))
    B.append(b)
B
```

```python
S = []
for t in tlist:
    SZW = np.sqrt(1/(1+np.cos(WV*(t-x))))*SZ
    s = SZW*SX 
    S.append(s)
S
```

```python
phi = []
for t in range(N):
    PHI = B[t]*S[t]
    phi.append(PHI)

dphi = []
for i in range(N-1):
    dphi.append(phi[i]-phi[i+1])
phi
dphi
```

```python
c = R/L
dV = []
for j in range(N-1):
    dv = c*dphi[j]
    dV.append(dv)
    
dV
```

```python
V = []
for l in range(N):
    if l == 0:
        V.append(V0)
    else:
        V.append(V[l-1]+dV[l-1])            
```

```python
V
```

```python
import matplotlib.pyplot as plt
plt.plot(tlist, V)
plt.show()
```
