```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from qutip import *
```

```python
#h = 10**-13 #10**-21 ampletude of gravitational wave
D = 10**3 #frequency
c = 3*10**8 #speed of light
G = 6.67384*10**-11
BX = 0.000025 #earth field is 25microtesla 
WV = D/c #wave vector
x = 0 #initial point
N = 20000 #no. of iterations
tlist = np.linspace(0, 0.1, N) #time list
SY = 0.001 #dimentions of squid in y dirction
SZ = 0.01 #dimentions of squid in z dirction
phi0 = 2.067833758*10**-15 #magnetic flux quanta
Ic = 1 #squid critical value
Ms = 1.989*10**30 #mass of sun
M = 2.4*Ms #mass of neutron star
R = 12*10**3 #radius of neutron star
ge = 1 #neutron star ellipticity
rs = 4*10**18 #distance to proxima centauri (ot any more)
#all in SI units
```

```python
#ampletude of GW 
h = (G*M*(R**2)*(D**2)*ge)/((c**4)*rs)
```

```python
#magnetic field in GW
B = []
for t in tlist:
    b = BX+1/2*BX*(h*WV*np.sin(WV*(t-x)))/(-1+h*np.cos(WV*(t-x)))
    B.append(b)
```

```python
#change squid area in GW
SW = []
S = SZ*SY
for t in tlist:
    sW = 2*h*np.sin(D*t)*SZ
    SW.append(sW)
    
```

```python
#magnetic flux in GW
phi = []
PN = []
PW = []
for t in range(N):
    PHIW = B[t]*SW[t]
    PHIN = B[t]*S
    pN = np.pi*(PHIN/phi0)
    pW = np.pi*(PHIW/phi0)
    PN.append(pN)
    PW.append(pW)
```

```python
#Current mesured by squid
I = []
for l in range(N):
    i = 2*Ic*abs(np.cos(PN[l]+PW[l]))
    I.append(i)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(figsize=(12,6))   # plot of expectation values
axes.plot(tlist, I)
axes.set_xlabel(r'$t/s$', fontsize=20)
axes.set_ylabel(r'$I/A$', fontsize=20)
```
